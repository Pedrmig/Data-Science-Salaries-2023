#### Preguntas a resolver usando las librerías de creaciones gráficas vistas en clase: 

¿Cuál es el salario promedio por nivel de experiencia?

```python
fig = px.bar(df, x='experience_level', y='salary_in_usd', template="presentation", title="Salary by Experience Level")
fig.show()
```

¿Cómo ha cambiado el salario medio con el paso del tiempo (work_year)?

```python
fig = px.bar(df, x='work_year', y='salary_in_usd', template="presentation", title="Salary by Work Year")
fig.show()
```

¿Cuál es la proporción de trabajo remoto entre diferentes roles de trabajo?

```python
px.pie(df, values='remote_ratio', names='employment_type', template="plotly_dark", title="Remote Ratio by Employment Type " ).show()
```

¿Existe una relación entre el tamaño de la empresa y el salario?

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['company_size'], df['salary_in_usd'])
plt.title('Corelacion entre tamaño de la empresa y el salario')
plt.xlabel('Tamaño da Empresa')
plt.ylabel('Salario')
plt.show()
```

```python
df.head()
```

¿Cuál es el país con más trabajadores remotos?

```python
df['remote_ratio'] = df['remote_ratio'].astype(bool)
df_remotos = df[df['remote_ratio'] == True].groupby('employee_residence').size().reset_index(name='counts')
fig = px.scatter(df_remotos, x='employee_residence', y='counts', color='employee_residence', title="Trabajadores Remotos por País")
fig.show()
```

```python
df.head()
```

¿Cómo ha cambiado la proporción de trabajo remoto con el paso del tiempo?

```python
df_remotos = df[df['remote_ratio'] == True].groupby('work_year').size().reset_index(name='counts')
fig = px.line(df_remotos, x="work_year", y="counts", title='Trabajadores Remotos por Año', template="plotly_dark") 
fig.show()
```

¿Cuál es el país con los salarios promedio más altos en USD?

```python
average_salary = df.groupby('employee_residence')['salary_in_usd'].mean()
country_with_highest_average_salary = average_salary.idxmax()
print(f'O país com o maior salário médio é: {country_with_highest_average_salary}')
```

```python
average_salary_df = pd.DataFrame(average_salary).reset_index()
average_salary_df.columns = ['employee_residence', 'salary_in_usd']
fig = px.bar(average_salary_df, x='employee_residence', y='salary_in_usd', template="plotly_dark")
fig.show()
```

¿Cómo se distribuyen los roles de trabajo en las diferentes categorías de experiencia?

```python
job_counts = df.groupby(['job_title', 'employment_type']).size().reset_index(name='counts')
job_counts_df = pd.DataFrame(job_counts)

fig = px.scatter(job_counts_df, 
                 x="job_title", 
                 y="counts", 
                 color="employment_type", 
                 hover_data=['employment_type'], 
                 log_x=False,
                 size_max=60)
fig.show()
```

##### Cómo varía el salario promedio entre los diferentes tipos de empleo, fulltime, partime ect?

```python
salary = df.groupby(['salary_in_usd', 'employment_type']).size().reset_index(name='counts')
salary_df = pd.DataFrame(salary)
fig = px.scatter(salary_df, x="salary_in_usd", y="counts", color="employment_type", hover_data=['employment_type'], size_max=60)
fig.show()
```

##### Crear un top10 con los empleos mas demandados:

```python
job_counts = df['job_title'].value_counts().reset_index()
job_counts.columns = ['job_title', 'counts']
top_10_jobs = job_counts.head(10)['job_title']
filtered_df = df.loc[df['job_title'].isin(top_10_jobs)]
filtered_df.loc[:, 'job_title_count'] = filtered_df['job_title'].map(filtered_df['job_title'].value_counts()) 
fig = px.bar(filtered_df, x='job_title', y='job_title_count', color='job_title')
fig.show()
```

#####  Crear un top con empleos donde podemos trabajar remoto: 

```python

remote = df.groupby(['job_title', df['remote_ratio'] == True]).size().reset_index(name='counts')
remote.columns = ['job_title', 'is_remote', 'counts']

filtered_df = remote.loc[remote['job_title'].isin(remote.head(10)['job_title'])]

fig = px.bar(filtered_df, x="job_title", y="counts", color="job_title")
fig.show()
```
